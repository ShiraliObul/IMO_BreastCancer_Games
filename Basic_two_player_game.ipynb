```python
# simulate

import matplotlib.pyplot as plt
%matplotlib inline

# payoffs a > c and b < d
#a = 2; b = 3; c = 1; d = 4
a = 4; b = 3; c = 1; d = 5

xA = [0.55]
xB = [1 - xA[0]]
dt = 0.1
# fitness of A and B
FA = [(xA[0] * a + xB[0] * b) * dt]
FB = [(xA[0] * c + xB[0] * d) * dt]
# average fitness
F = [(xA[0] * (xA[0] * a + xB[0] * b) + xB[0] * (xA[0] * c + xB[0] * d)) * dt]


for t in range(100):
    # fitnesses
    fA = xA[t] * a + xB[t] * b
    fB = xA[t] * c + xB[t] * d
    f = xA[t] * fA + xB[t] * fB
    FA.append(fA*dt)
    FB.append(fB*dt)
    F.append(f*dt)
    # differential equations
    xA.append(xA[t] + (xA[t] * (fA - f)) * dt)
    xB.append(xB[t] + (xB[t] * (fB - f)) * dt)

plt.plot(xA, 'r', label ='share of strategy A')
plt.plot(xB, 'b', label ='share of strategy B')
plt.plot(FA, 'r--', label ='fitness of strategy A')
plt.plot(FB, 'b--', label ='fitness of strategy B')
plt.plot(F, 'g--', label ='mean population fitness')
plt.grid()
plt.ylim(0, 1)
plt.legend(loc='best')
```
